##### Microtask-5

```python
X=df2[["btc_market_cap","btc_n_transactions","btc_miners_revenue","btc_cost_per_transaction","btc_difficulty","btc_hash_rate","btc_cost_per_transaction_percent"]]
X
```

```python
y=df2["btc_market_price"]
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)

from sklearn.linear_model import LinearRegression
model=LinearRegression()

train=model.fit(X_train,y_train)
```

```python
train.coef_
```

```python
train.intercept_
```

```python
train.score(X_train,y_train)
```

```python
pred=train.predict(X_test)
pred
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
MSE=mean_squared_error(y_test,pred)
MSE
```

```python
MAE=mean_absolute_error(y_test,pred)
MAE
```
